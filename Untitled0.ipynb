```python
# ==========================
# FINAL WORKING VERSION – November 20, 2025
# Fully free, no API keys, works for any ticker
# ==========================
!pip install yfinance --quiet   # run once in Colab

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================
# USER SETTINGS
# ==========================
SYMBOL = "AAPL"                  # Change to whatever you want
YEARS_OF_PRICES = 6
EVENT_PRE_DAYS  = 10
EVENT_POST_DAYS = 10
# ==========================

# ---- 1. Prices ----
ticker = yf.Ticker(SYMBOL)
hist = ticker.history(period=f"{YEARS_OF_PRICES}y", interval="1d")
if hist.empty:
    raise ValueError("No data — check ticker symbol")
df_prices = hist[["Close"]].rename(columns={"Close": "close"})

# ---- 2. Earnings dates (super robust) ----
earnings = ticker.get_earnings_dates(limit=50)   # most reliable endpoint

if earnings is None or earnings.empty:
    # fallback for some tickers
    earnings = ticker.quarterly_earnings

# Extract announcement dates
if earnings is not None and not earnings.empty:
    if "Earnings Date" in earnings.columns:
        ann_dates = pd.to_datetime(earnings["Earnings Date"]).dt.tz_localize(None)
    else:
        # quarterly_earnings uses index as quarter-end → use first trading day after
        ann_dates = []
        for qe in earnings.index:
            try:
                next_day = df_prices.index[df_prices.index > qe][0]
                ann_dates.append(next_day)
            except:
                pass
        ann_dates = pd.Series(ann_dates)
else:
    raise ValueError("No earnings data found for this ticker")

# Keep last 12 announcement dates
announcement_dates = sorted(ann_dates.dropna())[-12:]

print(f"Using last {len(announcement_dates)} earnings announcement dates:")
print([d.strftime("%Y-%m-%d") for d in announcement_dates])

# ---- 3. Long-term price chart with markers ----
plt.figure(figsize=(16,8))
plt.plot(df_prices.index, df_prices["close"], label="Close Price", color="#1f77b4")
for ed in announcement_dates:
    plt.axvline(ed, color="red", linestyle="--", alpha=0.7)
plt.title(f"{SYMBOL} Stock Price – Earnings Dates Marked")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ---- 4. Event study – average cumulative return ----
cum_rets = []
for ann in announcement_dates:
    try:
        idx = df_prices.index.get_loc(ann)
    except KeyError:
        continue
    if idx < EVENT_PRE_DAYS:
        continue

    window = df_prices.iloc[idx - EVENT_PRE_DAYS : idx + EVENT_POST_DAYS + 1].copy()
    window["day"] = np.arange(-EVENT_PRE_DAYS, EVENT_POST_DAYS + 1)
    base_price = window.iloc[0]["close"]
    window["cum_ret"] = (window["close"] / base_price - 1) * 100
    cum_rets.append(window.set_index("day")["cum_ret"])

if cum_rets:
    avg_ret = pd.concat(cum_rets, axis=1).mean(axis=1)
    plt.figure(figsize=(12,7))
    plt.plot(avg_ret.index, avg_ret.values, marker="o", color="#d62728", linewidth=2)
    plt.title(f"Average {SYMBOL} Return Around Earnings (±{EVENT_PRE_DAYS} trading days)")
    plt.xlabel("Trading Days Relative to Earnings (0 = announcement day)")
    plt.ylabel("Average Cumulative Return (%)")
    plt.axvline(0, color="black", linewidth=1)
    plt.axhline(0, color="black", linewidth=0.8)
    plt.grid(alpha=0.3)
    plt.show()
else:
    print("Not enough overlapping data for event study")

# ---- 5. Individual reactions table ----
rows = []
for ann in announcement_dates:
    try:
        idx = df_prices.index.get_loc(ann)
    except KeyError:
        continue
    if idx == 0:
        continue
    prev_close = df_prices.iloc[idx-1]["close"]
    r1 = (df_prices.iloc[idx  ]["close"] / prev_close - 1) * 100
    r2 = (df_prices.iloc[idx+1]["close"] / prev_close - 1) * 100 if idx+1 < len(df_prices) else np.nan
    r5 = (df_prices.iloc[idx+4]["close"] / prev_close - 1) * 100 if idx+4 < len(df_prices) else np.nan

    rows.append({
        "Date": ann.strftime("%Y-%m-%d"),
        "1-day %": round(r1, 2),
        "2-day %": round(r2, 2),
        "5-day %": round(r5, 2)
    })

print("\nIndividual earnings reactions:")
print(pd.DataFrame(rows).to_string(index=False))
```
