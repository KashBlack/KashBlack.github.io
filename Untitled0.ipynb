```python
import pandas as pd
import requests
import json
import sqlite3  # For SQL practice

# Step 1: Fetch the latest Retail Sales Index dataset CSV via ONS API
dataset_id = "retail-sales-index"
try:
    # Get dataset metadata
    response = requests.get(f"https://api.beta.ons.gov.uk/v1/datasets/{dataset_id}")
    response.raise_for_status()  # Raise error if not 200
    info = response.json()

    # Get latest version URL
    version_url = info['links']['latest_version']['href']
    version_response = requests.get(version_url)
    version_response.raise_for_status()
    version_info = version_response.json()

    # Get CSV download URL
    download_url = version_info['downloads']['csv']['href']
    print(f"Downloading from: {download_url}")

    # Load into DataFrame (skip metadata rows at top)
    df = pd.read_csv(download_url, skiprows=7)  # Adjust skiprows if needed after inspecting
    print("Data loaded successfully.")
    print("Column names:")
    print(df.columns)
    print("First few rows:")
    print(df.head())


except Exception as e:
    print(f"Error fetching data: {e}")
    print("If this fails, manually download the CSV from https://www.ons.gov.uk/businessindustryandtrade/retailindustry/datasets/retailsales and upload to Colab, then use df = pd.read_csv('your_file.csv', skiprows=7)")

# Step 2: Cleaning & Prep (assuming df is loaded)
if 'df' in locals():
    # Rename columns if needed (inspect df.columns; typical: 'Date', 'J3MC' for total volume, 'J4MC' for internet %)
    # Example: Filter for J4MC series (internet sales %)
    # Based on the head, 'Sep-25' seems to be the date column. Let's assume '106.5' is the value column for now.
    # We need to find the column that corresponds to J4MC. Looking at the ONS website, J4MC is 'Non-store retailing (internet sales) - Percentage of total retail sales for all retailing (including automotive fuel)'.
    # From the head, 'Non-store retailing' is in the column named 'Total of predominantly non-food stores'. Let's assume the value column is the first unnamed column '106.5'.
    df_j4mc = df[df['Total of predominantly non-food stores'] == 'Non-store retailing'].copy()
    df_j4mc = df_j4mc[['Sep-25', '106.5']].copy() # Re-select the relevant columns after filtering
    df_j4mc.columns = ['Date', 'J4MC'] # Rename columns for clarity
    df_j4mc['Date'] = pd.to_datetime(df_j4mc['Date'], format='%b-%y')
    df_j4mc['J4MC'] = pd.to_numeric(df_j4mc['J4MC'], errors='coerce')

    print("\ndf_j4mc after filtering and type conversion:")
    print(df_j4mc.head())
    print(df_j4mc.info())

    # Calculate growth (YoY % change)
    df_j4mc['growth'] = df_j4mc['J4MC'].pct_change(periods=12, fill_method=None) * 100
    df_j4mc.dropna(inplace=True)

    # Save for Power BI import
    df_j4mc.to_csv('uk_retail_sales_j4mc.csv', index=False)
    print("Cleaned data saved as 'uk_retail_sales_j4mc.csv'")

    # Step 3: SQL Example (in-memory DB for practice)
    conn = sqlite3.connect(':memory:')
    df_j4mc.to_sql('sales', conn, index=False)
    query = "SELECT strftime('%Y', Date) as year, AVG(growth) as avg_growth FROM sales GROUP BY year;"
    result = pd.read_sql(query, conn)
    print("SQL result:")
    print(result)
else:
    print("Data fetch failedâ€”check error above and try manual download.")
```

```python
# Query 1: Top growth years
query_top_growth = """
SELECT strftime('%Y', Date) as year, AVG(growth) as avg_growth
FROM sales
GROUP BY year
ORDER BY avg_growth DESC
LIMIT 5;
"""
top_growth = pd.read_sql(query_top_growth, conn)
print("Top growth years:")
print(top_growth)

# Query 2: Post-COVID recovery (2020+)
query_covid = """
SELECT Date, J4MC, growth
FROM sales
WHERE Date >= '2020-01-01'
ORDER BY Date;
"""
covid_recovery = pd.read_sql(query_covid, conn)
print("Post-COVID data:")
print(covid_recovery)
```
